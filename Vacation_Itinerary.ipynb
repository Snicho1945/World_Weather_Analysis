```python
df1 = pd.concat([Laguna_df, South_Lake_Tahoe_df])
df1
```

```python
df2 = pd.concat([df1, Fallon_df])
df2
```

```python
Travel_df = pd.concat([df2, Susanville_df])
Travel_df
```

```python
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Weather} and {Max Temp} °F</dd>
</dl>
"""
# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in Travel_df.iterrows()]
```

```python
import gmaps
import gmaps.datasets
gmaps.configure(api_key= g_key)
locations = Travel_df[["Lat", "Lng"]]

# Latitude-longitude pairs
Laguna = (38.42, -121.42)
South_Lake_Tahoe = (38.93, -119.98)
Fallon = (39.47, -118.78)
Susanville = (40.42, -120.65)

fig = gmaps.figure()
LagunaT = gmaps.directions_layer(Laguna, Susanville, waypoints=[South_Lake_Tahoe, Fallon], travel_mode='DRIVING')
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(marker_layer)
fig.add_layer(LagunaT)
fig
```
